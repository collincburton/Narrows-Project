```python
import requests
from bs4 import BeautifulSoup

def scan_sites():
    open_sites=[]
    site_ids=['30027001', '30027003', '30027005', '30027007', '30027009', '30027012', '40032002']
    page = requests.get("https://zionpermits.nps.gov/wilderness.cfm?TripTypeID=1")
    soup = BeautifulSoup(page.content, 'html.parser')
    for site in site_ids:
        tag=soup.find('option', value=site)
        if "znpwpresclosed" in str(tag):
            pass
        else:
            open_sites.append(tag.get_text())
    return open_sites
```
